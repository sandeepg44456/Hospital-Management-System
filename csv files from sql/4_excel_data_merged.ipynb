```python
import pandas as pd
```

```python
file1 = pd.read_csv('patients.csv')
file1
```

```python
file2 = pd.read_csv('province_names.csv')
file2
```

```python
merge_1 = pd.merge(file1, file2, on='province_id', how='inner')
merge_1
```

```python
file3 = pd.read_csv('admissions.csv')
file3
```

```python
file4 = pd.read_csv('doctors.csv')
file4
```

```python
merge_2 = pd.merge(file3, file4, left_on= 'attending_doctor_id', right_on='doctor_id', how='inner')
merge_2
```

```python
merge_3 = pd.merge(merge_1, merge_2, on ='patient_id', how= 'inner')
merge_3
```

```python
final_table = merge_3.drop(columns = ['doctor_id'])
final_table
```

```python
final_table.rename(columns={'first_name_x':'patient_first_name','last_name_x':'patient_last_name','first_name_y':'doctor_first_name','last_name_y':'doctor_last_name'}, inplace = 'TRUE')
```

```python
final_table
```

```python
combined_table = final_table[['patient_id','patient_first_name','patient_last_name','gender','birth_date',
                              'height','weight','city', 'province_id','province_name','allergies',
                              'admission_date','discharge_date','diagnosis',
                              'attending_doctor_id','doctor_first_name','doctor_last_name','specialty']]
```
