```python
import pandas as pd
```

```python
df = pd.read_csv('hospital_management_dataset.csv')
df
```

```python
df.info()
```

```python
df.columns
```

```python
### Q1: Show first name, last name, and gender of patients whose gender is 'M'.

df.loc[df['gender'] == 'M',['patient_first_name','patient_last_name','gender']].drop_duplicates().reset_index(drop=True)
```

```python
### Q2: Show first name and last name of patients that weight within the range of 100 to 120 (inclusive).

df.loc[(df['weight'] >=100) & (df['weight'] <= 120),['patient_first_name','patient_last_name','weight']].drop_duplicates().reset_index(drop=True)
```

```python
### Q3: Show first name of patients that start with the letter 'C'.

df[df['patient_first_name'].str.startswith('C')]['patient_first_name'].drop_duplicates().reset_index(drop=True)
```

```python
### Q4: Show first name and last name concatenated into one column to show their full name.

df['full_name'] = df['patient_first_name'] +' '+df['patient_last_name']
df['full_name'].drop_duplicates().reset_index(drop=True)
```

```python
### Q5: Show first name, last name, and the full province name of each patient. (Example: 'Ontario' instead of 'ON')

df[['patient_first_name','patient_last_name','province_name']].drop_duplicates().reset_index(drop=True)
```

```python
### Q6: Show how many patients have a birth_date with 2010 as the birth year.

df['birth_date'] = pd.to_datetime(df['birth_date'])

df[df['birth_date'].dt.year == 2010]['patient_id'].unique().shape[0]
```

```python
### Q7: Show the first_name, last_name, and height of the patient with the greatest height.

df.loc[df['height'].idxmax(),['patient_first_name','patient_last_name','height']]
```

```python
### Q8: Show all columns for patients who have one of the following patient_ids: 1,45,534,879,1000.

df[df['patient_id'].isin([1,45,534,879,1000])].drop_duplicates().reset_index(drop=True)
```

```python
### Q9: Show the total number of admissions.

df['admission_date'].count()
```

```python
### Q10: Show all the columns from admissions where the patient was admitted and discharged on the same day.

df[df['admission_date'] == df['discharge_date']].drop_duplicates().reset_index(drop=True)
```

```python
### Q11: Show the patient id and the total number of admissions for patient_id 579.

df[df['patient_id'] == 579].shape[0]
```

```python
### Q12: Based on the cities that our patients live in, show unique cities that are in province_id 'NS'?

df[df['province_id'] == 'NS']['city'].unique()
```

```python
### Q13: Write a query to find the first_name, last name and birth date of patients who has height greater than 160 and weight greater than 70.

df[(df['height'] > 160) & (df['weight'] >70)][['patient_first_name','patient_last_name','birth_date']].drop_duplicates().reset_index(drop=True)
```

```python
### Q14: Write a query to find list of patients first_name, last_name, and allergies where allergies are not null and are from the city of 'Hamilton'.


df[(df['allergies'].notnull()) & (df['city'] == 'Hamilton')][['patient_first_name','patient_last_name']].drop_duplicates().reset_index(drop=True)
```
